```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Change to your actual sequence path
seq_path = '/home/albert/Downloads/robopack-data/1_6_box5/train/seq_1'

# === Choose time index and cam/bubble index
t_index = '100'
cam_id = 0
bubble_id = 1

# === Load and visualize RGB-D camera
color = np.load(os.path.join(seq_path, f'cam_{cam_id}', f'color_{t_index}_{cam_id}.npy'))[0]
depth = np.load(os.path.join(seq_path, f'cam_{cam_id}', f'depth_{t_index}_{cam_id}.npy'))[0]

print(f'color shape: {color.shape}, depth shape {depth.shape}')

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(color)
plt.title(f'cam_{cam_id} RGB')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(depth, cmap='viridis')
plt.title(f'cam_{cam_id} Depth')
plt.axis('off')
plt.show()
```

```python

# === Load and visualize tactile sensors
bubble_force = np.load(os.path.join(seq_path, f'bubble_{bubble_id}', f'force_{t_index}_{int(t_index) // 100 - 1}.npy'))[0]
bubble_depth = np.load(os.path.join(seq_path, f'bubble_{bubble_id}', f'depth_{t_index}_{int(t_index) // 100 - 1}.npy'))[0]

# plt.figure(figsize=(10, 4))
# plt.subplot(1, 2, 1)
# plt.imshow(bubble_force, cmap='hot')
# plt.title(f'bubble_{bubble_id} Force')
# plt.axis('off')
print(f'force shape: {bubble_force.shape}, depth shape: {bubble_depth.shape}')

plt.subplot(1, 2, 2)
plt.imshow(bubble_depth, cmap='plasma')
plt.title(f'bubble_{bubble_id} Depth')
plt.axis('off')
plt.show()
```

```python
# visualize flow data


# === Load and visualize tactile sensors
bubble_flow = np.load(os.path.join(seq_path, f'bubble_{bubble_id}', f'raw_flow_{t_index}_{int(t_index) // 100 - 1}.npy'))[0]

bubble_flow = bubble_flow.reshape(320, 480)

# plt.figure(figsize=(10, 4))
# plt.subplot(1, 2, 1)
# plt.imshow(bubble_force, cmap='hot')
# plt.title(f'bubble_{bubble_id} Force')
# plt.axis('off')
print(f'flow shape: {bubble_flow.shape}')

plt.subplot(1, 2, 2)
plt.imshow(bubble_depth, cmap='plasma')
plt.title(f'bubble_{bubble_id} Flow')
plt.axis('off')
plt.show()
```
